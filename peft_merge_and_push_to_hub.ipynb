```python
from peft import AutoPeftModelForCausalLM, PeftModel
from transformers import AutoModelForCausalLM, AutoTokenizer
import torch
```

```python
from huggingface_hub import notebook_login
notebook_login()
```

```python
def merge_and_load(base_model_id, lora_adapter):
  base_model=AutoModelForCausalLM.from_pretrained(base_model_id, torch_dtype=torch.float16, load_in_8bit=False,
                                             device_map="auto",
                                             trust_remote_code=True)
  model=PeftModel.from_pretrained(base_model, lora_adapter, from_transformers=True, device_map="auto")
  merged_model=model.merge_and_unload()
  return merged_model
```

```python
#variables
model_id="mistralai/Mistral-7B-Instruct-v0.3"
lora_adapter='/content/drive/MyDrive/mistral-7b-eng-od-translate'

merged_model=merge_and_load(model_id, lora_adapter)
merged_model.save_pretrained('/content/drive/MyDrive/mistral-merged')
```

```python
tokenizer=AutoTokenizer.from_pretrained(model_id)
tokenizer.save_pretrained('/content/drive/MyDrive/mistral-merged')
```
